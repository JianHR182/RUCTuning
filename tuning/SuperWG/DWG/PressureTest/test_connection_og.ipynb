```python
import time
with open('workloads/sibench_workload.wg', 'r') as f:
    file_contents = f.read()

print(len(file_contents))
print(file_contents[0:100])

import re
file_contents_list=re.split(r'[;\n]+',file_contents)
print(len(file_contents_list))
for i,it in enumerate(file_contents_list):
    file_contents_list[i]+=";"
print(file_contents_list[0:5])
```

```python
start_time = time.time()
for i,it in enumerate(file_contents_list):
    cur.execute(it)
    connection.commit()
    if i%100==0:
        print(f"{i} sqls has been processed successfully.")
    
        end_time = time.time()
        elapsed_time = end_time - start_time
        print(f"目前所用时间：{elapsed_time}秒")
```
